```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取 CSV 数据文件
df = pd.read_csv('results_summary.csv')

# 为了更好地区分 TTP 与 KCTSP 数据，将两部分数据分开
df_ttp = df[df['problem'] == 'TTP']
df_kctsp = df[df['problem'] == 'KCTSP']

# 设置 Seaborn 样式
sns.set(style="whitegrid", font_scale=1.2)

#############################################
# 针对 TTP 问题的可视化
#############################################
# 绘制 TTP 各方法下不同实例的目标函数值对比图
plt.figure(figsize=(10, 6))
sns.barplot(data=df_ttp, x='instance', y='objective', hue='method', ci=None)
plt.title('Comparaison des valeurs objectives pour TTP')
plt.xlabel('Instance')
plt.ylabel('Valeur de l\'objectif')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/TTP_objective_comparison.png')
plt.show()

# 绘制 TTP 各方法下不同实例的计算时间（ms）对比图
plt.figure(figsize=(10, 6))
sns.barplot(data=df_ttp, x='instance', y='time_ms', hue='method', ci=None)
plt.title('Comparaison des temps de calcul pour TTP')
plt.xlabel('Instance')
plt.ylabel('Temps de calcul (ms)')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/TTP_time_comparison.png')
plt.show()

#############################################
# 针对 KCTSP 问题的可视化
#############################################
# 绘制 KCTSP 各方法下不同实例的目标函数值对比图
plt.figure(figsize=(10, 6))
sns.barplot(data=df_kctsp, x='instance', y='objective', hue='method', ci=None)
plt.title('Comparaison des valeurs objectives pour KCTSP')
plt.xlabel('Instance')
plt.ylabel('Valeur de l\'objectif')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/KCTSP_objective_comparison.png')
plt.show()

# 绘制 KCTSP 各方法下不同实例的计算时间（ms）对比图
plt.figure(figsize=(10, 6))
sns.barplot(data=df_kctsp, x='instance', y='time_ms', hue='method', ci=None)
plt.title('Comparaison des temps de calcul pour KCTSP')
plt.xlabel('Instance')
plt.ylabel('Temps de calcul (ms)')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/KCTSP_time_comparison.png')
plt.show()

#############################################
# 如果需要展示 PLNE_complete 的优化间隙 (gap) 对比（针对 KCTSP）
#############################################
# 过滤出 PLNE_complete 数据
df_plne_complete = df_kctsp[df_kctsp['method'] == 'PLNE_complete']
if not df_plne_complete.empty:
    plt.figure(figsize=(8, 5))
    sns.barplot(data=df_plne_complete, x='nb_items', y='gap', ci=None)
    plt.title('Optimisation gap pour PLNE_complete (KCTSP)')
    plt.xlabel('Nombre d\'objets')
    plt.ylabel('Gap')
    plt.tight_layout()
    plt.savefig('results/visualisation/KCTSP_PLNE_complete_gap.png')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取 CSV 数据文件
df = pd.read_csv('results_summary.csv')

# 将 time_ms 转为数字，如果有缺失则填充默认值（例如 300000 毫秒，用于超时情况）
df['time_ms'] = pd.to_numeric(df['time_ms'], errors='coerce')
df['time_ms'].fillna(300000, inplace=True)

# objective 也转换为数字（如果存在缺失，依据实际情况进行填充或剔除）
df['objective'] = pd.to_numeric(df['objective'], errors='coerce')

# 根据 problem 字段将数据分为 TTP 与 KCTSP 两部分
df_ttp = df[df['problem'] == 'TTP']
df_kctsp = df[df['problem'] == 'KCTSP']

# 设置 Seaborn 样式
sns.set(style="whitegrid", font_scale=1.2)

#############################################
# 针对 TTP 问题的可视化（使用 pointplot 展示变化趋势）
#############################################

# 目标函数值对比图（TTP）
plt.figure(figsize=(10, 6))
sns.pointplot(data=df_ttp, x='instance', y='objective', hue='method',
              markers=["o", "s", "D", "X"], linestyles='-', capsize=0.1)
plt.title('Comparaison des valeurs objectives pour TTP')
plt.xlabel('Instance')
plt.ylabel('Valeur de l\'objectif')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/TTP_objective_comparison_pointplot.png')
plt.show()

# 计算时间对比图（TTP）
plt.figure(figsize=(10, 6))
sns.pointplot(data=df_ttp, x='instance', y='time_ms', hue='method',
              markers=["o", "s", "D", "X"], linestyles='-', capsize=0.1)
plt.title('Comparaison des temps de calcul pour TTP')
plt.xlabel('Instance')
plt.ylabel('Temps de calcul (ms)')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/TTP_time_comparison_pointplot.png')
plt.show()

#############################################
# 针对 KCTSP 问题的可视化
#############################################

# 目标函数值对比图（KCTSP）
plt.figure(figsize=(10, 6))
sns.pointplot(data=df_kctsp, x='instance', y='objective', hue='method',
              markers=["o", "s", "D", "v", "^"], linestyles='-', capsize=0.1)
plt.title('Comparaison des valeurs objectives pour KCTSP')
plt.xlabel('Instance')
plt.ylabel('Valeur de l\'objectif')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/KCTSP_objective_comparison_pointplot.png')
plt.show()

# 计算时间对比图（KCTSP）
plt.figure(figsize=(10, 6))
sns.pointplot(data=df_kctsp, x='instance', y='time_ms', hue='method',
              markers=["o", "s", "D", "v", "^"], linestyles='-', capsize=0.1)
plt.title('Comparaison des temps de calcul pour KCTSP')
plt.xlabel('Instance')
plt.ylabel('Temps de calcul (ms)')
plt.legend(title='Méthode', loc='best')
plt.tight_layout()
plt.savefig('results/visualisation/KCTSP_time_comparison_pointplot.png')
plt.show()

#############################################
# 针对 KCTSP 的 PLNE_complete 部分，展示优化间隙 (gap)
#############################################
df_plne_complete = df_kctsp[df_kctsp['method'] == 'PLNE_complete']
if not df_plne_complete.empty:
    plt.figure(figsize=(8, 5))
    # 将 nb_items 转为数值类型便于排序（如果需要）
    df_plne_complete['nb_items'] = pd.to_numeric(df_plne_complete['nb_items'], errors='coerce')
    sns.pointplot(data=df_plne_complete.sort_values('nb_items'), x='nb_items', y='gap', markers="o", capsize=0.1)
    plt.title('Optimisation gap pour PLNE_complete (KCTSP)')
    plt.xlabel('Nombre d\'objets')
    plt.ylabel('Gap')
    plt.tight_layout()
    plt.savefig('results/visualisation/KCTSP_PLNE_complete_gap_pointplot.png')
    plt.show()
```
